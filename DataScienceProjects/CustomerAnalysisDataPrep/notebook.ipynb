A common problem when creating models to generate business value from data is that the datasets can be so large that it can take days for the model to generate predictions. Ensuring that your dataset is stored as efficiently as possible is crucial for allowing these models to run on a more reasonable timescale without having to reduce the size of the dataset.

You've been hired by a major online data science training provider called *Training Data Ltd.* to clean up one of their largest customer datasets. This dataset will eventually be used to predict whether their students are looking for a new job or not, information that they will then use to direct them to prospective recruiters.

You've been given access to `customer_train.csv`, which is a subset of their entire customer dataset, so you can create a proof-of-concept of a much more efficient storage solution. The dataset contains anonymized student information, and whether they were looking for a new job or not during training:

| Column                   | Description                                                                      |
|------------------------- |--------------------------------------------------------------------------------- |
| `student_id`             | A unique ID for each student.                                                    |
| `city`                   | A code for the city the student lives in.                                        |
| `city_development_index` | A scaled development index for the city.                                         |
| `gender`                 | The student's gender.                                                            |
| `relevant_experience`    | An indicator of the student's work relevant experience.                          |
| `enrolled_university`    | The type of university course enrolled in (if any).                              |
| `education_level`        | The student's education level.                                                   |
| `major_discipline`       | The educational discipline of the student.                                       |
| `experience`             | The student's total work experience (in years).                                  |
| `company_size`           | The number of employees at the student's current employer.                       |
| `company_type`           | The type of company employing the student.                                       |
| `last_new_job`           | The number of years between the student's current and previous jobs.             |
| `training_hours`         | The number of hours of training completed.                                       |
| `job_change`             | An indicator of whether the student is looking for a new job (`1`) or not (`0`). |

```python
# Import necessary libraries
import pandas as pd

# Load the dataset
ds_jobs = pd.read_csv("customer_train.csv")

# View the dataset
ds_jobs.head()
```

```python
# Create a copy of ds_jobs for transforming
ds_jobs_transformed = ds_jobs.copy()

# Start coding here. Use as many cells as you like!
```

```python
#review the dtypes and info on them
print(ds_jobs_transformed.info())

#review for potential unqie values
print(ds_jobs_transformed.isna().value_counts())
```

```python
#lets determine which cat features have which
ordinal_cols = ['education_level', 'experience', 'company_size', 'last_new_job', 'enrolled_university']

for col in ordinal_cols:
    print(ds_jobs_transformed[col].unique())
```

```python
#create the ordered lists for ordinal data types

edu_order = ['Primary School', 'High School', 'Graduate', 'Masters', 'Phd']

#order exp by val
#since <1 and >20 are not digits, we need to specify them, let the for loop complete the middle
exp_order = ['<1'] + [str(i) for i in range(1, 21)] + ['>20']

comp_size_order = ['<10', '10-49', '50-99', '100-499', '500-999', '1000-4999', '5000-9999', '10000+']

lnj_order = ['never', '1', '2', '3', '4', '>4']

uni_order = ['no_enrollment', 'Part time course', 'Full time course']
```

```python
#category map for easier use with astype (since it accepts dicts)

cat_map = {
    'education_level': pd.CategoricalDtype(edu_order, ordered=True),
    'experience': pd.CategoricalDtype(exp_order, ordered=True),
    'company_size': pd.CategoricalDtype(comp_size_order, ordered=True),
    'last_new_job': pd.CategoricalDtype(lnj_order, ordered=True),
    'enrolled_university': pd.CategoricalDtype(uni_order, ordered=True)
    
}

#fuse them into the df
ds_jobs_transformed = ds_jobs_transformed.astype(cat_map)

print(ds_jobs_transformed[ordinal_cols].info())
```

```python
#lets do nominal cats next (that are not bool)
nominal_cols = ['city', 'gender', 'major_discipline', 'company_type']

ds_jobs_transformed[nominal_cols] = ds_jobs_transformed[nominal_cols].astype('category')

print(ds_jobs_transformed[nominal_cols].info())
```

```python
#let's start by filtering out data types

#locate cols where it can be bool
boolVals = ds_jobs_transformed.columns[ds_jobs_transformed.nunique() == 2]

#locate bit ints next
intVals = ds_jobs_transformed.select_dtypes(include = ['int']).columns
intVals = intVals.difference(boolVals)

#locate floats
floatVals = ds_jobs_transformed.select_dtypes(include = ['float']).columns
#filter out bools 
floatVals = floatVals.difference(boolVals)
```

```python
#create a dict to store keys for transformation

type_dict = {col: 'bool' for col in boolVals}
type_dict.update({col: 'int32' for col in intVals})
type_dict.update({col: 'float16' for col in floatVals})

#fuse
ds_jobs_transformed = ds_jobs_transformed.astype(type_dict)
```
